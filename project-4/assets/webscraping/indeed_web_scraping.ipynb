## Web-Scraping Indeed.com
This script is created to scrape relevant jobs from indeed.com, using the libraries, Selenium and bs4

```python
import numpy as np
import pandas as pd

from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.alert import Alert

from time import sleep

import html2text
```

```python
h = html2text.HTML2Text()
h.ignore_links = True
```

```python
job_data = []
```

```python
for title in ['data analyst','data scientist', 'business intelligence data']:
    
    url = ('https://www.indeed.com.sg/')
    driver = webdriver.Chrome(executable_path='/Users/user/Documents/GA-DSI/course-materials/classes/week-06/labs/python-webscraping_opentable-lab-master/chromedriver/chromedriver_1')

    driver.get(url)

    ## Keying in search term

    search_term = driver.find_element_by_id('text-input-what')
    search_term.send_keys(title)

    sleep(np.random.randint(1,5))

    driver.find_element_by_xpath('//*[@id="whatWhere"]/form/div[3]/button').click()
    default_handle=driver.current_window_handle


    for page in np.arange(10):
        print 'this is page ', page
        
        for i in np.arange(0,15):
            print i
            this_job = {}
            sleep(np.random.randint(1,4))
            try:
                driver.find_element_by_class_name('popover-x-button-close').click()   ## in case of popup dialog, this will close box 
            except:
                pass

            try:
                try:
                    driver.find_elements_by_class_name('jobtitle')[i].click()
                    print i,'job click suc'
                except:
                    print i,'job click fail'
                    continue
                driver.switch_to_window(driver.window_handles[1])

                sleep(np.random.randint(2,4))

                            ## Extracting the data required

                HTML = driver.page_source
                soup = BeautifulSoup(HTML,'lxml')
                try:
                    this_job['job_title'] = soup.find('b',{'class':'jobtitle'}).text ## Job title
                    print i,' title ext suc'
                except:    
                    this_job['job_title'] = soup.find('h3',{'class':'jobsearch-JobInfoHeader-title'}).text
                    print i,' title ext suc'
                try:
                    this_job['salary'] = soup.find('span',{'class':'no-wrap'}).text ## Salary
                    print i,' salary ext suc'
                except:
                    pass
                this_job['company'] = soup.find('span',{'class':'company'}).text ## Company
                print i,' comp ext suc'
                this_job['jd'] = soup.find('span',{'id':'job_summary'}).text ## Getting the job description
                print i,' jd ext suc'

                job_data.append(this_job)

                sleep(np.random.randint(3,5))

                            ## Close current listed job page and return to main page with all the other listings

                driver.close()
                driver.switch_to_window(default_handle)
                print i            

            except:
                print 'error',i


        ## Goto next page
        try:
            driver.find_elements_by_class_name('np')[-1].click()
            print 'next page',page
        except:
            print 'next page fail',page
            driver.close()
            break
```

```python
df = pd.DataFrame(job_data)
```
