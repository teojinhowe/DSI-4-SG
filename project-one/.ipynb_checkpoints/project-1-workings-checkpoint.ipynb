### Creating function to make new Pokedex

```python
def cre_new_pokedex():
    
    stat_names = cleanraw[0][1:]     ## Creates a list of the stat names(i.e keys is listed dictionary in individual pokemon stat dictionary)
                                    ## PokedexNumber is left out
    pokemon_data = cleanraw[1:]    ## Takes out the pokemon data (pokemon_id included), creates a list of list of Pokemon and their stats
    
    new_pokedex_dict = {}         ## dictionary Pokemon_id with dictionary of Pokemon stats
    
    for sel_pokemon in pokemon_data:    ## Iterating through Pokemon_data, individual Pokemon
        
        sel_pokemon_stats = {}              ## Preparing a stats dictionary for particular Pokemon in this iteration 
        
        for i,stat_name in enumerate(stat_names):           ## Matching key(which is stat_name) to value(which is the corresponding value in pokemon data)
            sel_pokemon_stats[stat_name] = sel_pokemon[i+1]        ## [i+1] to align corresponding value and name. 
                                                                    ## Refer to comment on stats_names and pokemon_data creation
        
        new_pokedex_dict[sel_pokemon[0]] =sel_pokemon_stats  ## Matching a key value pair with the pokemon_id as key, dict for stats as value
    
    return new_pokedex_dict
```

```python
new_pokedex = cre_new_pokedex()
```

### Finding pop mean and std dev of "Total" attribute

```python
import numpy as np

total_list = [pokemon['Total'] for pokemon in new_pokedex.values()]
total_mean = np.mean(total_list)
total_std = np.std(total_list)

print('"Total" mean: {}'.format(total_mean))
print('"Total" std dev: {}'.format(total_std))
```

### Probability of pokemon appearance

```python
### using z score to reverse rank. low z score = higher probability to appear. high z score = lower probability to appear
```

```python
for n in new_pokedex.values():
    n['Total_z'] = (n['Total']-total_mean)/total_std
    
```

```python
## Creating a list of pokemon sorted based on Total_z score

##Part 1 - creating the new list to be sorted later
z_ranked_pokedex = []
for k,v in new_pokedex.items():
    x = {}
    x['pokemon_id'] = k
    x['Name'] = v['Name']
    x['Total_z'] = v['Total_z']
    
    z_ranked_pokedex.append(x)
```
