## Statistics
### 1.What are mean, median, mode, standard deviation of [8,4,2,1,4,5]?
    Mean = 4
    Median = 4
    Mode = 4
    Standard Deviation = 2.45
### 2.What is a distribution? What is a normal distribution??
    A distribution is a listing or function showing all the possible values or intervals (bins) of the data and 
    how often they occur in a dataset.
    A normal distribution is a distribution that is symmetrical, where the mean and median are usually equal.
### 3.What is the 25th percentile of a distribution?
    The 25th percentile is the first quartile, which has the smallest value that is greater or equal to the lowest 25% of the
    all the values in the dataset.
### 4.What is the interquartile range
    The range between the 75th percentile and 25th percentile.

```python
import numpy as np
from scipy import stats
import pandas as pd

qns_value = [8,4,2,1,4,5]
print "mean: {}".format(np.mean(qns_value))
print "median: {}".format(np.median(qns_value))
print "mode: {}".format(stats.mode(qns_value))
print "std: {}".format(np.std(qns_value,ddof=1))
print stats.describe(qns_value)
```
